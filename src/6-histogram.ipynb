### Exercise 6: Histogram Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values from 0 to 10
x = np.arange(0, 11, 1)
# Compute y values as the square of x
y = x ** 2

# Data for bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']

# Data for scatter plot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

# Data for pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# First subplot: Line plot
axes[0, 0].plot(x, y, marker='o', linestyle='-', color='b', label='y = x^2')
axes[0, 0].set_title('Line Plot of y = x^2')
axes[0, 0].set_xlabel('X Values')
axes[0, 0].set_ylabel('Y Values (X Squared)')
axes[0, 0].grid(True)
axes[0, 0].legend()

# Second subplot: Bar plot
axes[0, 1].bar(categories, values, color=colors)
axes[0, 1].set_title('Bar Plot of Categories')
axes[0, 1].set_xlabel('Categories')
axes[0, 1].set_ylabel('Values')

# Third subplot: Scatter plot
axes[1, 0].scatter(x_scatter, y_scatter, color='magenta')
for i, txt in enumerate(zip(x_scatter, y_scatter)):
    axes[1, 0].annotate(f'({txt[0]}, {txt[1]})', (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(5,5), ha='center')
axes[1, 0].set_title('Scatter Plot with Annotations')
axes[1, 0].set_xlabel('X Values')
axes[1, 0].set_ylabel('Y Values')
axes[1, 0].grid(True)

# Fourth subplot: Pie chart
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['gold', 'skyblue', 'lightcoral', 'lightgreen'], startangle=140)
axes[1, 1].set_title('Pie Chart of Programming Languages')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()

# Data for histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Create the histogram
plt.figure(figsize=(6,4))
plt.hist(data, bins=5, color='teal', edgecolor='black')

# Customize the histogram
plt.title('Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the histogram
plt.show()
```
