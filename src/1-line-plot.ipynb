### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values from 0 to 10
x = np.arange(0, 11, 1)
# Compute y values as the square of x
y = x ** 2

# Create the plot
plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = x^2')

# Customize the plot
plt.title('Line Plot of y = x^2')
plt.xlabel('X Values')
plt.ylabel('Y Values (X Squared)')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
